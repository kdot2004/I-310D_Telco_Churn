# Drop the columns that won't be considered as features

```python
df.head()
```

```python
#Check number of unique values in each column
df.nunique()
```

**Explanation of columns dropped:**
1. customer_id: The unique identifier of each customer, serves no predictive value.
2. count: Another unique identifier used for dashboarding, not useful for prediction. 
3. country: Given that all the customers are from the same country there's no variability meaning it's not very useful for prediciton. 
4. city: Given the high amount of unique cities represented in the dataset, it'd be best to drop this column rather than one-hot encoding it.
5. zip_code: For both ethical considerations and model simplicity I think it'd be best to drop the zip_codes column. For ethical considerations, zip codes can perpetuate socio-economic and potential racial biases. Thus if the model determines that certain zip codes lead to higher likelihoods of churn this might reflect biases that are better off avoided. For model simplicity because of the high amount of unique values of zip codes, despite being a numerical feature, could still add uneccessary complexitity to the model that'll be better off avoided. Although I could just use the first 2-3 integers of the zip codes to reduce it's cardinality, it'll probably just be better to avoid zip codes.
6. lat_long, latitude, longitude: Similarly to zip codes, these location based features will be dropped due to their high number of unique values and the potential biases that come along with these. 
7. churn_label: This column is redundant since we have churn_value, which is it's binary representation. 
8. churn_score: Churn score isn't considered a explanatory variable, especially since it's derived from another ML model that incorporates other criteria. Using it would defeat the purpose of training our own model. 
9. cltv: Similarly to churn_score cltv is better left off because it's not truly a explanatory variable (It's derived from internal business models that assign importance to each customer). 

10. churn_reason: Again this isn't truly a explanatory variable and is more so a afterprouduct of churn. Were after what factors lead up to churn, with the goal of preventing it before it happens. 

**More Potential Columns to drop:**
- Gender: Gender is a protected class and additionally could produce biased results in ML models. Currently we'll keep this feature but in the future we'll test the model without it. From our EDA we already saw that it didn't seem highly correlated with churn, meaning that one gender didn't have a significantly higher proportion of churn compared to the other. 
- Senior Citizen Status: Again age is a protected class. senior_citizen represents customers who are 65 years or older. Age is also a protected class and can produce biases within ML models. From our EDA we also saw a pretty big difference in proportion of churns between customers who are considered senior citizens and those who aren't. Thus we will keep this feature for now but it will be tested without it in the future to test for fariness.

# Encoding Categorical Features

```python
#Make a copy of orignal dataframe for potential future use
original_df = df.copy()

#Encode current dataset, use drop_first for binary columns
df = pd.get_dummies(df, columns=['gender', 'senior_citizen', 'partner', 'dependents', 'phone_service', 'paperless_billing'], drop_first=True)

#For multi-category columns I'll choose which columns to manually drop
df = pd.get_dummies(df, columns=['multiple_lines', 'internet_service',
       'online_security', 'online_backup', 'device_protection', 'tech_support',
       'streaming_tv', 'streaming_movies', 'contract', 'payment_method'])

#Ensure all columns are still lowercase
df.columns = df.columns.str.lower()
```

```python
#Replace spaces with underscores
df.columns = df.columns.str.replace(' ', '_')
```

```python
#Print out columns
df.columns
```

```python
#I wanted to choose which multi-category columns to drop based off which one's had the weakest correlation with churn (from the EDA)

#For multiple line status I'll drop `multiple_lines_no_phone_service`
#For internet service I'll drop  `internet_service_no`
#For online security I'll drop `online_security_no_internet_service`
#For online backup I'll drop `online_backup_no_internet_service`
#For device protection I'll drop `device_protection_no_internet_service`
#For tech support I'll drop `tech_support_no_internet_service`
#For streaming tv I'll drop `streaming_tv_no_internet_service`
#For streaming movies I'll drop `streaming_movies_no_internet_service`
#For contract I'll drop `contract_two_year`
#For payment method I'll drop `payment_method_credit_card_(automatic)`

df = df.drop(columns=['multiple_lines_no_phone_service', 'internet_service_no', 'online_security_no_internet_service',
            'online_backup_no_internet_service', 'device_protection_no_internet_service', 'tech_support_no_internet_service',
            'streaming_tv_no_internet_service', 'streaming_movies_no_internet_service', 'contract_two_year', 
            'payment_method_credit_card_(automatic)'])

#Change the payment method column for easier handling
df.rename(columns={
    'payment_method_bank_transfer_(automatic)': 'payment_method_bank_transfer' 
},inplace=True)

#Display head
df.head()
```

```python
#Convert to binary columns to 1s and 0s (Uneccessary but a prefference thing)
def convert_bools_to_ints(df):
    bool_cols = df.select_dtypes(include='bool').columns
    df[bool_cols] = df[bool_cols].astype(int)
    return df
df = convert_bools_to_ints(df)

#Check result
df.head()
```

```python
#Re-order columns
order = ['tenure_months', 'monthly_charges', 'total_charges', 'gender_male', 'senior_citizen_yes', 'partner_yes', 'dependents_yes',
       'phone_service_yes', 'paperless_billing_yes', 'multiple_lines_no',
       'multiple_lines_yes', 'internet_service_dsl',
       'internet_service_fiber_optic', 'online_security_no',
       'online_security_yes', 'online_backup_no', 'online_backup_yes',
       'device_protection_no', 'device_protection_yes', 'tech_support_no',
       'tech_support_yes', 'streaming_tv_no', 'streaming_tv_yes',
       'streaming_movies_no', 'streaming_movies_yes',
       'contract_month-to-month', 'contract_one_year',
       'payment_method_bank_transfer', 'payment_method_electronic_check',
       'payment_method_mailed_check', 'churn_value']
df = df[order]

#Check result
df.head()
```

```python
#Save both dataframes to csv file
original_df.to_csv('telco_churn.csv',index=False) #Original dataframe containing only necessary columns
df.to_csv('telco_churn_encoded.csv',index=False) #Encoded dataframe
```

# Checking for High-Correlation

```python
#Import data viz libraries
import matplotlib.pyplot as plt
import seaborn as sns

#Create correlation matrix
corr_matrix = df.corr(numeric_only=True)

#Visualize the correlation matrix as a heatmap
plt.figure(figsize=(20,18)) #Set figure size
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)
```

**Breakdown:**
- Tenure Months and Total Charges: Strong linear relationship (r=0.83), makes sense as we saw from our EDA that customers who've stayed with Telco longer have total higher charges. Obviously as time moves forward then tenure months increases. 
- Monthly Charges and Internet Service Fiber Optic: Strong linear relationship (r=0.79), customers with fiber optic internet (1) tend to have higher monthly charges, while those without it (0) tend to have lower charges.
- Monthly Charges and Streaming Services: Customers who pay for streaming TV or streaming movies tend to have higher monthly charges which is  logical because these are likely add-on services (r=0.63).
- Monthly Charges and Total Charges: Naturally, someone with high monthly charges over time will accumulate high total charges (r=0.65). I might drop `total_charges` in the future because its unnecessary since we have `tenure_months`. Total Charges = Monthly Charges x Tenure Months (Not including additional charges). 
- Total Charges and Add-On Services: Customers who pay for add-on services (streaming, device protection, etc) tend to have higher total charges which is  logical because these are likely add-on services.
- Streaming TV and Movies: Moderate linear relationship (r=0.54), customers who subscribe to streaming TV are somewhat likely to also subscribe to streaming movies, and vice versa but not always.

**Application:**
Will test my models without `total_charges` in the future because I get can a similar effect from `tenure_months`. Additionally `total_charges` is somewhat a combination of `monthly_charges` and `tenure_months`. 

# Normalizing Features: Logistic Regression and Naive Bayes

The assumptions of both Logistic Regression and Naive Bayes is that numerical features as normally distributed. We have three numerical  features, `tenure_months`, `monthly_charges` and `total_charges`. First we'll explore the current distribution of the data, and then apply the necessary transformation to normalize it. 

```python
#Import data viz libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Make a histogram of tenure_months
plt.figure(figsize=(8,6)) #Set figure size
sns.histplot(x='tenure_months', data=df, kde=True)
plt.title('Distribution of Tenure Months')
plt.xlabel('Tenure Months')
plt.show()
```

```python
#Make a histogram of monthly_charges
plt.figure(figsize=(8,6)) #Set figure size
sns.histplot(x='monthly_charges', data=df, kde=True)
plt.title('Distribution of Monthly Charges')
plt.xlabel('Monthly Charges')
plt.show()
```

```python
#Make a histogram of total_charges
plt.figure(figsize=(8,6)) #Set figure size
sns.histplot(x='total_charges', data=df, kde=True)
plt.title('Distribution of Total Charges')
plt.xlabel('Total Charges')
plt.show()
```

```python
#Check the skewness of each
df[['tenure_months', 'monthly_charges', 'total_charges']].skew()
```

```python
#Check the minimum value of each of our numerical columns
df[['tenure_months', 'monthly_charges', 'total_charges']].min()
```

```python
#Apply log transformation to total_charges
import numpy as np #Import numpy
df['log_total_charges'] = np.log(df['total_charges'])

#Check distribution again
sns.histplot(x='log_total_charges', data=df, kde=True)
plt.title('Distribution of Total Charges Following Log Transformation')
plt.xlabel('Log Total Charges')
plt.show()
```

```python
#Check skewness of total_charges
df['log_total_charges'].skew()
```

```python
#Apply different transformations to total_charges to view if we can get a better result
df['sqrt_total_charges'] = np.sqrt(df['total_charges']) #Sqaure root
df['cbrt_total_charges'] = np.cbrt(df['total_charges']) #Cube root

#Import scipy for more advanced transformations
from scipy import stats 
df['boxcox_total_charges'],_ = stats.boxcox(df['total_charges']) #Boxcox transformation
df['yj_total_charges'],_ = stats.yeojohnson(df['total_charges']) #Yeo Johnson transformation

#Check result
for col in ['sqrt_total_charges', 'cbrt_total_charges', 'boxcox_total_charges', 'yj_total_charges']:
    print(f"Skewness for {col}: {df[col].skew()}")
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
#Given that the cbrt is the closest to 0, I'll drop the rest of the transformed columns
df.drop(columns=['yj_total_charges', 'boxcox_total_charges', 'sqrt_total_charges', 'log_total_charges'], inplace=True)
```

```python
#Check the distribution of Churn Value
sns.histplot(data=df, x='monthly_charges', hue='churn_value', kde=True)
plt.title("Monthly Charges by Churn")
plt.show()
```

# Scaling Features: K-NN and SVM

For distance based algorithms such as K-Nearest Neighbors and Support Vector Machines the distance or the range of numerical features matter. The furhter apart data points are from each other, the more it'll effect the models performance. This means the presence of outliers in any of our numerical features can drastically effect the model's performance. As such for K-NN and SVM, we'll apply min-max scaling to change the range of the data to zeros and ones. 

```python
#Import pandas and read in encoded telco churn dataframe
import pandas as pd
df = pd.read_csv('telco_churn_encoded.csv')
df.head()
```

```python
#Import min-max scaling from scikit-learn
from sklearn.preprocessing import MinMaxScaler

#Initialize list of columns to scale
scaled_cols = ['tenure_months', 'monthly_charges', 'total_charges']

#Initialize scaler
scaler = MinMaxScaler()

#Make a copy of the dataframe
scaled_df = df.copy()

#Apply the scaler to the scaled dataframe
scaled_df[scaled_cols] = scaler.fit_transform(scaled_df[scaled_cols])

#Check result
scaled_df[['tenure_months', 'monthly_charges', 'total_charges']].head()
```

```python
#Save to a csv file
scaled_df.to_csv('knn_svm_scaled_telco_churn.csv',index=False)
```
