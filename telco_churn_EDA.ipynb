```python
#Visualize result
plt.figure(figsize=(8,6)) #Set figure size
plot = sns.barplot(y='churn_reason', x='count', data=churn_reasons, palette='rocket') #Create plot
#Annotate plot
plt.title('Popular Churn Reasons amongst Former Telco Customers')
plt.xlabel('Frequency')
plt.ylabel('Churn Reason')
for i in plot.containers: #Add value counts to each churn reason
    plot.bar_label(i, )
plt.show() #display plot
```

```python
#Filter dataset to remove "Unknown" from churn reason
known_churns = df[df['churn_reason'] != 'Unknown']

#Apply function to classify churn reasons for more specific focus
def specify_churn(x):
    if 'Competitor' in x:
        return 'Competitor Related'
    elif any(keyword in x for keyword in ['Attitude', 'reliability', 'dissatisfaction', 'services', 'Poor', 'expertise']):
        return 'Poor Service'
    elif any(keyword in x for keyword in ['Price too high', 'Long distance', 'Extra', 'Lack of']):
        return 'Financial/Price Concerns'
    elif any(keyword in x for keyword in ['Moved', 'Deceased']):
        return 'Client-Specific'
    else:
        return 'Unknown' 

#Apply function to filtered dataframe
known_churns['specific_churn_reason'] = known_churns['churn_reason'].apply(specify_churn)
```

```python
#Check value counts
known_churns['specific_churn_reason'].value_counts()
```

```python
#Visualize result
churns = known_churns['specific_churn_reason'].value_counts().reset_index() #Get the counts
plt.figure(figsize=(8,6)) #Set figure size
plot = sns.barplot(x='specific_churn_reason', y='count', data= churns, palette='rocket')
#Annotate
plt.title('Common reasons for Churns Amongst Customers')
plt.xlabel('Churn Reason')
plt.xticks(rotation=25)
plt.ylabel('Frequency')
for i in plot.containers: #Add counts
    plot.bar_label(i, )
#Display
plt.show()
```

## Total Charges

### Filling Null Values in Total Charges

```python
#Check descriptive statistics
print('Customers who churned')
print(churn['total_charges'].describe())
print('\n')
print("Customers who havn't churned")
print(non_churn['total_charges'].describe())
```

Hm that's interesting, it should display descriptive statistics, lets check the data type of `total_charges` to see if it's a numeric.

```python
#Check data type of total_charges
df['total_charges'].dtype
```

Since it's a object column I'll need to convert it to a float, using the `astype` method. 

```python
#Then let's convert it to a float
df['total_charges'] = df['total_charges'].astype(float)
```

The error we got indicates that there's missing data in the `total_charges` column. The missing data is currently empty, as shown by the ''. To fix this we need to replace these empty values with some sort of number, I'll use 0 in this case. Afterwards I'll convert `total_charges` to a float and then I'll replace the zeros with the median or the mode of the total_charges depending on if it's normally distributed. However, let's first check if there's already zero's present in `total_charges`. If not then we'll proceed. 

```python
#Check for zero's in `total_charges`
df[df['total_charges'] == '0']
```

Good since, there's no zero's I can go ahead and convert the empty strings to zero's, then check the distribution to impute the missing data. 

```python
#First let's check which specific indices have a empty string for `total_charges`
df[df['total_charges'] == ' '].index
```

I was having some issues using the replace method on the entire column so I decided it'd be best to just call the specific indices with empty strings and manually change the values to zero's. 

```python
#Manually replace all of the empty strings for the indices with zero's 
df.loc[2234, 'total_charges'] = '0'
df.loc[2438, 'total_charges'] = '0'
df.loc[2568, 'total_charges'] = '0'
df.loc[2667, 'total_charges'] = '0'
df.loc[2856, 'total_charges'] = '0'
df.loc[4331, 'total_charges'] = '0'
df.loc[4687, 'total_charges'] = '0'
df.loc[5104, 'total_charges'] = '0'
df.loc[5719, 'total_charges'] = '0'
df.loc[6772, 'total_charges'] = '0'
df.loc[6840, 'total_charges'] = '0'
```

```python
#Check result
df[df['total_charges'] == '0']
```

```python
#Convert to flaot
df['total_charges'] = df['total_charges'].astype(float)
```

```python
#Convert zero's to null using numpy, (this makes it easier to impute)
import numpy as np #Import numpy 
df['total_charges'] = df['total_charges'].replace(0, np.nan)
```

```python
#Check the distribution of the data
plt.figure(figsize=(8,6))
sns.histplot(x='total_charges', data=df, kde=True)
plt.title('Distribution of Total Charges')
plt.show()
```

Since the distribution is not normal, I'll use median to impute the missing values. Additionally, since there's a small amount of missing values for `total_charges`, using median is great in this scenario. 

```python
#Find the median of total_charges
total_charges_median = df['total_charges'].median()
total_charges_median
```

```python
#Impute using total_charges_median
df['total_charges'] = df['total_charges'].fillna(total_charges_median)
```

# Summary of EDA: What makes a Customer Churn?

The analysis of customer churn for a fictional telecommunications company providing home phone and internet services reveals several key factors influencing customer retention. The churn rate in the dataset is 26.5%, with 1,869 out of 7,043 customers leaving the company. Demographically, females have a slightly higher churn rate (26.92%) compared to males (26.16%), and senior citizens exhibit a significantly higher churn rate (41.68%) compared to non-senior citizens (23.61%). Customers with dependents are less likely to churn, with a churn rate of 6.52%, while those without dependents experience a higher churn rate of 32.55%.

Having a partner also appears to play a notable role in customer retention. Customers without a partner have a significantly higher churn rate (32.96%) compared to those with a partner (19.66%). This trend suggests that customers with stronger household or family ties may have more stable service needs and are thus less likely to switch providers.

Service features also play a critical role in churn. Customers who subscribe to tech support have a significantly lower churn rate (15.17%) compared to those who do not have tech support (41.64%). This highlights the importance of providing strong customer service as a means to retain customers. Additionally, contract type is a key factor: customers on month-to-month contracts have the highest churn rate (42.71%), while those with two-year contracts have the lowest churn rate (2.83%). This suggests that longer contract terms correlate with lower churn, likely due to the commitment involved. In terms of internet service, fiber optic customers experience the highest churn rate (41.89%), which could indicate dissatisfaction or competition from other service providers, while customers with DSL have the lowest churn rate (18.96%).

Financial factors are also significant in understanding churn. Churned customers tend to have higher monthly charges on average ($74.44) compared to non-churned customers ($61.27). This could indicate that some customers are dissatisfied with the cost relative to the value they receive. Similarly, churned customers have lower total charges on average ($1,531.80) compared to non-churned customers ($2,552.88), suggesting that those who have been with the company longer, accumulating more charges, are less likely to leave.

We also categorized churn reasons, with poor service, competitor offers, and price concerns being the most common causes. For example, 704 customers left due to poor service, while 621 churned because of better offers from competitors, such as faster download speeds or more data. Price-related issues led to 331 customers leaving the company.

In conclusion, this analysis reveals that factors such as tech support, contract duration, payment method, pricing, and service offerings significantly impact customer churn. Demographic variables such as age, family status, and partnership also correlate with churn behavior. By improving customer service, offering longer contract terms, targeting high-risk demographics, and addressing competitive pressures, the company could reduce churn and improve customer retention.
